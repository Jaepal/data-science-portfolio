```python
import pandas as pd
```

```python
air = pd.read_csv('../data/gyeonggi_air.csv', encoding='euc-kr')
```

```python
air.drop(['아황산가스농도값(ppm)', '일산화탄소농도값(ppm)', '오존농도값(ppm)', '이산화질소농도값(ppm)'], axis='columns', inplace=True)
air.dropna(axis=0, inplace=True)
air.head()
```

```python
air2 = air.groupby(['측정일시각', '시군명'])['미세먼지PM10농도값(μg/m³)', '미세먼지PM2.5농도값(μg/m³)'].mean().reset_index()
air2
```

```python
def pm10_status(df):
    df.loc[ df['미세먼지PM10농도값(μg/m³)'] <= 30, '미세먼지PM10상태'] = '좋음'
    df.loc[(df['미세먼지PM10농도값(μg/m³)'] > 30) & (df['미세먼지PM10농도값(μg/m³)'] <= 80), '미세먼지PM10상태'] = '보통'
    df.loc[(df['미세먼지PM10농도값(μg/m³)'] > 80) & (df['미세먼지PM10농도값(μg/m³)'] <= 150), '미세먼지PM10상태'] = '나쁨'
    df.loc[ df['미세먼지PM10농도값(μg/m³)'] > 150, '미세먼지PM10상태'] = '매우나쁨'
```

```python
def pm25_status(df):
    df.loc[ df['미세먼지PM2.5농도값(μg/m³)'] <= 15, '미세먼지PM2.5상태'] = '좋음'
    df.loc[(df['미세먼지PM2.5농도값(μg/m³)'] > 15) & (df['미세먼지PM2.5농도값(μg/m³)'] <= 35), '미세먼지PM2.5상태'] = '보통'
    df.loc[(df['미세먼지PM2.5농도값(μg/m³)'] > 35) & (df['미세먼지PM2.5농도값(μg/m³)'] <= 75), '미세먼지PM2.5상태'] = '나쁨'
    df.loc[ df['미세먼지PM2.5농도값(μg/m³)'] > 75, '미세먼지PM2.5상태'] = '매우나쁨'
```

```python
pm10_status(air2)
pm25_status(air2)
air2.head()
```

```python
air2.to_csv('../data/gyeonggi_air_status.csv', index=False, encoding='euc-kr')
```
